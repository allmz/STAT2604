```python

import pandas as pd
import numpy as np

df.loc[(df['best_of'] == 3) & (df['minutes'].isna()), 'minutes'] = 90

df.loc[(df['best_of'] == 5) & (df['minutes'].isna()), 'minutes'] = 165
```

```python
df.columns.values
```

```python
djokovic_matches = df[(df['winner_name'] == 'Novak Djokovic') | (df['loser_name'] == 'Novak Djokovic')]

def match_lenght(row):
    if row['best_of'] == 3:
        return 'short' if row['minutes'] < 90 else ('long' if row['minutes'] > 100 else 'avarage')
    elif row['best_of'] == 5:
        return 'short' if row['minutes'] < 165 else ('long' if row['minutes'] > 175 else 'avarage')


djokovic_matches['match_length_category'] = djokovic_matches.apply(match_lenght, axis=1)
```

```python
prob_short_best_of_3 = len(djokovic_matches[(djokovic_matches['best_of'] == 3) & (djokovic_matches['match_length_category'] == 'short')]) / len(djokovic_matches[djokovic_matches['best_of'] == 3])
prob_long_best_of_3 = len(djokovic_matches[(djokovic_matches['best_of'] == 3) & (djokovic_matches['match_length_category'] == 'long')]) / len(djokovic_matches[djokovic_matches['best_of'] == 3])

prob_short_best_of_5 = len(djokovic_matches[(djokovic_matches['best_of'] == 5) & (djokovic_matches['match_length_category'] == 'short')]) / len(djokovic_matches[djokovic_matches['best_of'] == 5])
prob_long_best_of_5 = len(djokovic_matches[(djokovic_matches['best_of'] == 5) & (djokovic_matches['match_length_category'] == 'long')]) / len(djokovic_matches[djokovic_matches['best_of'] == 5])
```

```python
import random

def monte_carlo_simulation(num_simulations, prob_short, prob_long):
    short_count = 0
    long_count = 0

    for _ in range(num_simulations):
        rand_prob = random.random()  

        if rand_prob < prob_short:
            short_count += 1
        elif rand_prob < prob_short + prob_long:
            long_count += 1

    prob_short_simulated = short_count / num_simulations
    prob_long_simulated = long_count / num_simulations

    return prob_short_simulated, prob_long_simulated

num_simulations = 100000
short_long_probs_best_of_3 = monte_carlo_simulation(num_simulations, prob_short_best_of_3, prob_long_best_of_3)
short_long_probs_best_of_5 = monte_carlo_simulation(num_simulations, prob_short_best_of_5, prob_long_best_of_5)

print("Best of 3 - Short Match Winning Probability:", short_long_probs_best_of_3[0])
print("Best of 3 - Long Match Probability:", short_long_probs_best_of_3[1])
print("Best of 5 - Short Match Probability:", short_long_probs_best_of_5[0]) 
print("Best of 5 - Long Match Probability:", short_long_probs_best_of_5[1])

```
