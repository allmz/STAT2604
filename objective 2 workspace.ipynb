```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("concat.csv")
```

```python
df
```

```python
# using binary method of defining win/lose in match

djokovic_matches = df[(df['winner_name'] == 'Novak Djokovic') | (df['loser_name'] == 'Novak Djokovic')]

djokovic_matches['win'] = np.where(djokovic_matches['winner_name'] == 'Novak Djokovic', 1, 0)
```

```python
# wins
djokovic_matches.loc[djokovic_matches['win'] == 1, 'ace'] = djokovic_matches['w_ace']
djokovic_matches.loc[djokovic_matches['win'] == 1, 'df'] = djokovic_matches['w_df']
djokovic_matches.loc[djokovic_matches['win'] == 1, 'svpt'] = djokovic_matches['w_svpt']
djokovic_matches.loc[djokovic_matches['win'] == 1, 'bpSaved'] = djokovic_matches['w_bpSaved']
djokovic_matches.loc[djokovic_matches['win'] == 1, 'bpFaced'] = djokovic_matches['w_bpFaced']
djokovic_matches.loc[djokovic_matches['win'] == 1, 'rank'] = djokovic_matches['winner_rank']
djokovic_matches.loc[djokovic_matches['win'] == 1, 'age'] = djokovic_matches['winner_age']
```

```python
# loses
djokovic_matches.loc[djokovic_matches['win'] == 0, 'ace'] = djokovic_matches['l_ace']
djokovic_matches.loc[djokovic_matches['win'] == 0, 'df'] = djokovic_matches['l_df']
djokovic_matches.loc[djokovic_matches['win'] == 0, 'svpt'] = djokovic_matches['l_svpt']
djokovic_matches.loc[djokovic_matches['win'] == 0, 'bpSaved'] = djokovic_matches['l_bpSaved']
djokovic_matches.loc[djokovic_matches['win'] == 0, 'bpFaced'] = djokovic_matches['l_bpFaced']
djokovic_matches.loc[djokovic_matches['win'] == 0, 'rank'] = djokovic_matches['loser_rank']
djokovic_matches.loc[djokovic_matches['win'] == 0, 'age'] = djokovic_matches['loser_age']
```

```python
correlation_matrix = djokovic_matches[['ace', 'df', 'svpt', 'bpSaved', 'bpFaced', 'rank', 'age', 'win']].corr()

# View the correlation with the 'win' column
print(correlation_matrix['win'])
```

```python
import seaborn as sns
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
```
